```python
import numpy as np
from scipy.stats import norm, logistic, laplace, hypsecant
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 200)

plt.plot(x, norm.pdf(x, loc=0, scale=1),
        'r-',lw = 2,  alpha=0.6, label='Normal pdf')
plt.plot(x, logistic.pdf(x, loc = 0, scale = np.sqrt(3)/np.pi),
        'b-',lw = 2,  alpha=0.6, label='Logistic pdf')
plt.plot(x, laplace.pdf(x, loc = 0, scale = 1/np.sqrt(2)),
        'g-',lw = 2,  alpha=0.6, label='Laplace pdf')
plt.plot(x, hypsecant.pdf(x, loc = 0, scale = 1/np.sqrt(2.4674011002723395)),
        'm-',lw = 2,  alpha=0.6, label='Hypsecant pdf')
plt.title('Comparing Simple Noises\n  (mean =0 and variance = 1)')
plt.legend()
plt.savefig('Images/simple_noises_comparison', bbox_inches='tight')
plt.show()
```
